```python
# fazer o jogo parar quando a palavra for descoberta
# ou quando as tentativas do usuário esgotarem
```

```python
# forca

palavra = 'uva'.lower()
letras_acertadas = ['_' for letra in palavra]

enforcou = False
acertou = False

erros = 0

print(letras_acertadas)

while not enforcou and not acertou:
  chute = input('qual letra?\n').strip().lower()

  if chute in palavra:
    index = 0
    for letra in palavra:
      if chute == letra:
        letras_acertadas[index] = letra
      index += 1
  else:
    erros += 1
    print(f'ops, você errou! faltam {6-erros} tentativas')

  enforcou = erros == 6
  acertou = '_' not in letras_acertadas

  # if erros == 6:
  #   break
  # if '_' not in letras_acertadas:
  #   break

  print(letras_acertadas)

if acertou:
  print('ganhou')
else:
  print('perdeu')

print('fim do jogo')
```

Bloco 7<br>
Melhorando o código e a apresentação

```python
import random
```

```python
def imprimir_abertura():

  print('bem-vindo à forca')
```

```python
def carregar_palavra():

  palavras = []

  with open('palavras.txt', 'r') as arquivo:
    for linha in arquivo:
      palavras.append(linha.strip().lower())

  return palavras[random.randrange(0, len(palavras))]
```

```python
def inicializar_tracinhos(palavra):

  return ['_' for letra in palavra]
```

```python
def pedir_chute():
  return input('qual letra?\n').strip().lower()
```

```python
def marcar_correto(chute, tracinhos, palavra):
  index = 0
  for letra in palavra:
    if chute == letra:
      tracinhos[index] = letra
    index += 1
```

```python
def desenhar_forca(erros):
  print("  _______     ")
  print(" |/      |    ")

  if erros == 1:
    print(" |      (_)   ")
    print(" |            ")
    print(" |            ")
    print(" |            ")

  if erros == 2:
    print(" |      (_)   ")
    print(" |      \     ")
    print(" |            ")
    print(" |            ")

  if erros == 3:
    print(" |      (_)   ")
    print(" |      \|    ")
    print(" |            ")
    print(" |            ")

  if erros == 4:
    print(" |      (_)   ")
    print(" |      \|/   ")
    print(" |            ")
    print(" |            ")

  if erros == 5:
    print(" |      (_)   ")
    print(" |      \|/   ")
    print(" |       |    ")
    print(" |            ")

  if erros == 6:
    print(" |      (_)   ")
    print(" |      \|/   ")
    print(" |       |    ")
    print(" |      /     ")

  if erros == 7:
    print(" |      (_)   ")
    print(" |      \|/   ")
    print(" |       |    ")
    print(" |      / \   ")

  print(" |            ")
  print("_|___         ")
  print()
```

```python
def jogar():

  imprimir_abertura()

  palavra = carregar_palavra()

  tracinhos = inicializar_tracinhos(palavra)

  enforcou = False
  acertou = False
  erros = 0

  while not enforcou and not acertou:
    print(tracinhos)
    chute = pedir_chute()

    if chute in palavra:
      marcar_correto(chute, tracinhos, palavra)
    else:
      erros += 1
      print(f'ops, você errou! faltam {7-erros} tentativas')
      desenhar_forca(erros)

    enforcou = erros == 7
    acertou = '_' not in tracinhos

  if acertou:
    print('você ganhou')
    print(tracinhos)
  else:
    print('você perdeu')
    print(f'a palavra era {palavra}')
```

```python
jogar()
```
