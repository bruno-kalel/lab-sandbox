Bloco 4 (quatro)<br>
A sequência do jogo

```python
# permitir que o usuário final teste várias vezes
```

```python
numero_secreto = 42
```

```python
numero_tentativas = 3
```

```python
while numero_tentativas > 0:
  print('rodadas restantes', numero_tentativas)

  chute = int(input('chute um número secreto\n'))
  print('você digitou', chute)

  acertou = chute == numero_secreto
  maior = chute > numero_secreto
  menor = chute < numero_secreto

  if acertou:
    print('acertou')
    break
  else:
    if maior:
      print('seu chute foi MAIOR que o número secreto')
    elif menor:
      print('seu chute foi MENOR que o número secreto')

    numero_tentativas -= 1

print('fim do jogo')
```

```python
# formatação de strings
# newer
print(f'{numero_secreto}, {numero_tentativas}')
# older
print('{}, {}'.format(numero_secreto, numero_tentativas))
```

```python
# começamos com 1000 pontos
# perdemos o valor distância entre o número secreto e chute em pontos
```

```python
pontuacao = 1000
```

```python
print('escolha um nível válido')
nivel = 0
while nivel not in (1, 2, 3):
  nivel = int(input('1 - fácil (5t) / 2 - médio (4t) / 3 - difícil (3t)\n'))
if nivel == 1:
  numero_tentativas = 5
elif nivel == 2:
  numero_tentativas = 4
elif nivel == 3:
  numero_tentativas = 3
```

```python
import random
numero_secreto = random.randint(1, 50)
```

```python
for rodada in range(numero_tentativas):
  print('rodada', rodada+1)

  chute = int(input('chute um número secreto\n'))
  print('você digitou', chute)

  if not 0 < chute < 51:
    print('chute entre 1 e 50')
    continue

  acertou = chute == numero_secreto
  maior = chute > numero_secreto
  menor = chute < numero_secreto

  if acertou:
    print('acertou')
    break
  else:
    # se fosse usado a função abs(), cálculo de pontos poderia ir aqui
    # isto é, apenas um cálculo para o if e para o elif
    if maior:
      print('seu chute foi MAIOR que o número secreto')
      pontuacao -= chute - numero_secreto
      # pontuacao -= abs(chute - numero_secreto)
    elif menor:
      print('seu chute foi MENOR que o número secreto')
      # + com - = -
      pontuacao += chute - numero_secreto
      # pontuacao -= abs(chute - numero_secreto)

print('pontuação', pontuacao)
print('resposta', numero_secreto)
print('fim do jogo')
```
