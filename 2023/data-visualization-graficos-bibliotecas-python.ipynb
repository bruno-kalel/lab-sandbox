```python
import pandas as pd
```

```python
df = pd.read_csv('/content/imigrantes_canada.csv')
```

```python
df
```

```python
df.info()
```

```python
df.set_index('País', inplace=True)
```

```python
df
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
anos
```

```python
brasil = df.loc['Brasil', anos]
```

```python
brasil
```

```python
brasil_dict = {'ano':brasil.index.tolist(),
               'imigrantes':brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 4))

plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])

plt.title('imigração do brasil para o canadá')
plt.xlabel('ano')
plt.ylabel('número de imigrantes')

plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.yticks([500, 1000, 1500, 2000, 2500, 3000])

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax.set_title('imigração do brasil para o canadá\n1980 a 2013')
ax.set_xlabel('ano')
ax.set_ylabel('número de imigrantes')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 6))

axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('imigração do brasil para o canadá\n1980 a 2013')
axs[0].set_xlabel('ano')
axs[0].set_ylabel('número de imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('boxplot da imigração do brasil para o canadá\n1980 a 2013')
axs[1].set_xlabel('brasil')
axs[1].set_ylabel('número de imigrantes')
axs[1].grid()

plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6))
fig.subplots_adjust(hspace=0.4, wspace=0.4)
fig.suptitle('imigração dos 4 maiores países da américa do sul para o canadá\n'
             '1980 a 2013')

axs[0, 0].plot(df.loc['Brasil', anos])
axs[0, 0].set_title('brasil')

axs[0, 1].plot(df.loc['Colômbia', anos])
axs[0, 1].set_title('colômbia')

axs[1, 0].plot(df.loc['Argentina', anos])
axs[1, 0].set_title('argentina')

axs[1, 1].plot(df.loc['Peru', anos])
axs[1, 1].set_title('peru')

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('ano')
  ax.set_ylabel('número de imigrantes')

ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

plt.show()
```

```python
américa_sul = df.query('Região == "América do Sul"')
```

```python
américa_sul
```

```python
cores = ['royalblue',
         'orange',
         'forestgreen',
         'orchid',
         'purple',
         'brown',
         'slateblue',
         'gray',
         'olive',
         'navy',
         'teal',
         'tomato']

with plt.style.context('dark_background'):
  fig, ax = plt.subplots(figsize=(12, 5))
  ax.barh(américa_sul.index,
          américa_sul['Total'],
          color=cores)
  ax.set_title('imigração da américa do sul para o canadá\n1980 a 2013',
               loc='left',
               fontsize=14)
  ax.set_xlabel('número de imigrantes', fontsize=12)
  ax.set_ylabel('')
  ax.xaxis.set_tick_params(labelsize=10)
  ax.yaxis.set_tick_params(labelsize=10)

  plt.show()
```

```python
américa_sul_sorted = américa_sul.sort_values('Total', ascending=True)
```

```python
cores = []

for país in américa_sul_sorted.index:
  if país == 'Brasil':
    cores.append('green')
  else:
    cores.append('silver')

with plt.style.context('dark_background'):
  fig, ax = plt.subplots(figsize=(10, 4))
  ax.barh(américa_sul_sorted.index,
          américa_sul_sorted['Total'],
          color=cores)
  ax.set_title('américa do sul: brasil foi o 4º país com mais imigrantes\n'
  'para o canadá no período de 1980 a 2013',
               loc='left',
               fontsize=14)
  ax.set_xlabel('número de imigrantes', fontsize=12)
  ax.set_ylabel('')
  ax.xaxis.set_tick_params(labelsize=10)
  ax.yaxis.set_tick_params(labelsize=10)

  for i, v in enumerate(américa_sul_sorted['Total']):
    ax.text(v + 200,
            i,
            str(v),
            color='white',
            fontsize=10,
            ha='left',
            va='center')

  ax.set_frame_on(False)
  ax.get_xaxis().set_visible(False)
  ax.tick_params(axis='both',
                 which='both',
                 length=0)
  fig.savefig('1.png',
              transparent=False,
              dpi=300,
              bbox_inches='tight')
  plt.show()
```

```python
def p(palette):
  fig, ax = plt.subplots(figsize=(8, 4))
  ax = sns.barplot(data=top_10,
                   x='Total',
                   y=top_10.index,
                   orient='h',
                   palette=palette)
  ax.set_title('países com maior imigração para o canadá\n1980 a 2013',
               fontsize=14,
               loc='left')
  ax.set_xlabel('número de imigrantes',
                fontsize=12)
  ax.set_ylabel('')
  plt.show()
```
