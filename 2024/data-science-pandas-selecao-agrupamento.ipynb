###### Unindo dados

```python
dados_censo = '/content/drive/MyDrive/POP2022_Municipios.xls'
```

```python
pop_estados = pd.read_excel(dados_censo, header=1, skipfooter=34)
pop_estados.head()
```

```python
pop_estados.groupby('UF').sum(numeric_only=True).head()
```

```python
pop_estados[pop_estados['POPULAÇÃO'].str.contains('\(', na=False)]
```

```python
pop_estados = pop_estados.assign(pop_sem_p = pop_estados['POPULAÇÃO'].replace('\(\d{1,2}\)', '', regex=True),
                                 pop = lambda x: x.loc[:, 'pop_sem_p'].replace('\.', '', regex=True))
```

```python
pop_estados[pop_estados['POPULAÇÃO'].str.contains('\(', na=False)]
```

```python
pop_estados = pop_estados.astype({'pop' : 'int64'})
pop_estados.head()
```

```python
pop_estados = pop_estados.groupby('UF')[['pop']].sum().reset_index()
pop_estados.head()
```

```python
emissão_uf = emissões_melted[emissões_melted['Ano'] == 2021].groupby('Estado')[['Emissão']].sum().reset_index()
emissão_uf.head()
```

```python
dados_agrupados = pd.merge(emissão_uf, pop_estados, left_on='Estado', right_on='UF')
dados_agrupados.head()
```

```python
dados_agrupados.rename(columns={'pop' : 'População'}, inplace=True)
dados_agrupados.head()
```

```python
dados_agrupados.plot(x='População', y='Emissão', kind='scatter', figsize=(6, 6))
```
