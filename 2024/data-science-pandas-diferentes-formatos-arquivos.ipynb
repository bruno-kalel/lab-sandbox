#### Manipulando arquivos JSON

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/pacientes.json'
```

```python
pacientes = pd.read_json(url)
pacientes
```

```python
url_nested = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/pacientes_2.json'
```

```python
pacientes_nested = pd.read_json(url_nested)
pacientes_nested
```

```python
pacientes_normalized = pd.json_normalize(pacientes_nested['Pacientes'])
pacientes_normalized
```

#### Lendo dados em HTML e XML

```python
import requests

url = "https://en.wikipedia.org/wiki/AFI's_100_Years...100_Movies"

response = requests.get(url)

if response.status_code == 200:
    with open('webpage.html', 'wb') as f:
        f.write(response.content)
    print("Webpage saved as webpage.html")
else:
    print("Failed to fetch the webpage")
```

```python
dados = pd.read_html('webpage.html')
dados
```

```python
type(dados)
```

```python
top = dados[1]
top
```

```python
top.to_html('top.html')
!ls
```

```python
pd.read_html('top.html')
```

```python
top.to_csv('top.csv', index=False)
!ls
```

```python
pd.read_csv('top.csv')
```
